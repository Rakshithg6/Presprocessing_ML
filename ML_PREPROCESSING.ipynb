# Label encoding

```python
from sklearn.preprocessing import LabelEncoder

columns_to_encode = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
                      'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                      'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
                      'PaperlessBilling', 'PaymentMethod']

label_encoder = LabelEncoder()

for column in columns_to_encode:
    df[column] = label_encoder.fit_transform(df[column])
df
```

```python
# Dropping 'customerID' from the dataframe
df.drop('customerID', axis=1, inplace=True)
df
```

```python
df.info()
```

'TotalCharges' column has empty strings. Hence need to convert to numeric value.

```python
# Replace empty strings in 'TotalCharges' with NaN
df['TotalCharges'] = df['TotalCharges'].replace('', pd.NA)

# Convert 'TotalCharges' to numeric, handling errors and converting NaN to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Handle missing values in 'TotalCharges' (for example, by imputing with mean or median)
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)
```

```python
df.info()
```

```python
# Oversampling
from imblearn.over_sampling import RandomOverSampler

# Oversample the minority class
oversampler = RandomOverSampler()
X_resampled, y_resampled = oversampler.fit_resample(X_scaled, y)
```

# Train test split

```python
from sklearn.model_selection import train_test_split
# Splitting training and test set
X_train_resampled, X_test_resampled, y_train_resampled, y_test_resampled = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=1)

# Checking split of data
print('X training data size: {}'.format(X_train_resampled.shape))
print('y training data size: {}'.format(y_train_resampled.shape))
print('X testing data size:  {}'.format(X_test_resampled.shape))
print('y testing data size:  {}'.format(y_test_resampled.shape))
```
